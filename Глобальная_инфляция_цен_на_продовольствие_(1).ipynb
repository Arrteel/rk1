# *Вопрос 1:* "Какие страны лидируют по инфляции в мире?"

*Этот код строит гистограмму инфляции для первых 10 стран из набора данных.*

```python
global_inflation = df.groupby('country')['Inflation'].mean().sort_values(ascending=False).head(10)
```

```python
# диаграмма с matplotlib

df.dropna(subset=['Inflation'], inplace=True)
df['date'] = pd.to_datetime(df['date'])

df = df[df['Inflation'] >= 0]

selected_countries = df['country'].unique()[:10]

df_selected = df[df['country'].isin(selected_countries)]

plt.figure(figsize=(15, 12))
sns.histplot(data=df_selected, x="Inflation", kde=True, hue='country', palette='tab10', multiple='stack')
plt.title('Гистограмма инфляции по странам (первые 10)')
plt.show()
```

Гистограмма показывает 10 лидирующих стран по инфляции в мире. Это такие страны как: Афганистан, Бурунди, Буркина-Фасо, Республика центральной Африки, Камерун, Конго, Гамбия, Гвинея-Бисау и Гаити.

```python
# диаграмма с plotly

fig = px.bar(global_inflation, x=global_inflation.index, y=global_inflation.values,
             title='Топ-10 стран по средней инфляции', labels={'x': 'Страна', 'y': 'Средняя инфляция'})
fig.show()
```

# Вопрос 5: Какие страны испытывают наибольшую и наименьшую глобальную инфляцию цен на продовольствие?

```python
highest_inflation_countries = df.groupby('country')['Inflation'].mean().sort_values(ascending=False)
lowest_inflation_countries = df.groupby('country')['Inflation'].mean().sort_values()
```

```python
# matplotlib

fig, axes = plt.subplots(nrows=2, figsize=(12, 12))
sns.barplot(x=highest_inflation_countries.head(5).index, y=highest_inflation_countries.head(5).values, ax=axes[0])
axes[0].set_title('Топ-5 стран по средней инфляции цен на продовольствие')
axes[0].set_xlabel('Страна')
axes[0].set_ylabel('Средняя инфляция цен на продовольствие')

sns.barplot(x=lowest_inflation_countries.head(5).index, y=lowest_inflation_countries.head(5).values, ax=axes[1])
axes[1].set_title('Топ-5 стран по наименьшей инфляции цен на продовольствие')
axes[1].set_xlabel('Страна')
axes[1].set_ylabel('Средняя инфляция цен на продовольствие')

plt.tight_layout()
plt.show()
```

```python
# plotly

fig = px.bar(highest_inflation_countries.head(5), x=highest_inflation_countries.head(5).index,
             y=highest_inflation_countries.head(5).values,
             title='Топ-5 стран по средней инфляции цен на продовольствие', labels={'x': 'Страна', 'y': 'Средняя инфляция'})
fig.show()

fig = px.bar(lowest_inflation_countries.head(5), x=lowest_inflation_countries.head(5).index,
             y=lowest_inflation_countries.head(5).values,
             title='Топ-5 стран по наименьшей инфляции цен на продовольствие',
             labels={'x': 'Страна', 'y': 'Средняя инфляция'})
fig.show()
```
